Attribute Information:
a)sepal length in cm
b)sepal width in cm
c)petal length in cm
d)petal width in cm

# Label Encoding

Every machine learning model takes numerical features as input. In our dataset we have 3 labels in the species column.Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
```

```python
df.head()
```

```python
df['Species'].unique()
```

# Model Training

```python
x=df.drop(columns=['Species'])
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
# model training
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print(pd.DataFrame(y_pred))
```

```python
print(y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
neighbors=[1,2,3,4,5,6,7,8,9]
cv_score=[]
for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    scores=cross_val_score(knn,x_train,y_train,cv=3,scoring='accuracy')
    cv_score.append(scores.mean())
plt.plot(neighbors,cv_score)
plt.xlabel('Number of neighbors')
plt.ylabel('accuracy')
plt.show()
```

```python
#n=3 gives the highest accuracy
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
print(y_pred)
```

```python
print(y_test.values)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
depth=[1,2,3,4,5,6,7,8,9]
cv_score=[]
for d in depth:
    dt=DecisionTreeClassifier(max_depth=d)
    scores=cross_val_score(dt,x_train,y_train,cv=3,scoring='accuracy')
    cv_score.append(scores.mean())
plt.plot(neighbors,cv_score)
plt.xlabel('Depth of Tree')
plt.ylabel('Accuracy')
plt.show()
```

```python
dt=DecisionTreeClassifier(max_depth=4)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
