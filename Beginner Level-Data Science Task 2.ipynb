```python
#LSTM is very sensitive to scale of data. So therefore applying MinMaxScaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_high=scaler.fit_transform(df_high.values.reshape(-1,1))
```

```python
df_high
```

```python
#Train-Test Split:-
train_size=int(len(df_high)*0.75)
test_size=len(df_high)-train_size
```

```python
print(train_size)
print(test_size)
```

```python
train_df=df_high[:train_size,:]
test_df=df_high[train_size:,:1]
```

# Data Prepration

```python
#100 previous data points are stored
def create_dataset(data,timestep=1):
    data_x,data_y=[],[]
    for i in range(len(data)-timestep-1):
        a=data[i:(i+timestep),0]
        data_x.append(a)
        data_y.append(data[i+timestep,0])
    return np.array(data_x),np.array(data_y)
```

```python
timestep=100
x_train,y_train=create_dataset(train_df,timestep)
```

```python
x_test,y_test=create_dataset(test_df,timestep)
```

```python
#Now making a 3-D requirement for LSTM (samples,timesteps,features)
x_train=x_train.reshape(x_train.shape[0],x_train.shape[1],1)
x_test=x_test.reshape(x_test.shape[0],x_test.shape[1],1)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM

model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50,batch_size=64,verbose=1)
```

```python
train_pred=model.predict(x_train)
test_pred=model.predict(x_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
rmse_train=sqrt(mean_squared_error(y_train,train_pred))
rmse_test=sqrt(mean_squared_error(y_test,test_pred))
```

```python
print(rmse_train)
```

```python
print(rmse_test)
```

# Plots

```python
train_pred=scaler.inverse_transform(train_pred)
test_pred=scaler.inverse_transform(test_pred)
```
